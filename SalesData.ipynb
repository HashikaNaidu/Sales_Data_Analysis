```python

# Box Plot for Unit Cost by Age Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Unit_Cost', data=sales)
plt.title('Unit Cost Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Unit Cost')
plt.xticks(rotation=45)
plt.show()

# Bar Chart for Average Profit by Country
avg_profit_by_country = sales.groupby('Country')['Profit'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Profit', data=avg_profit_by_country)
plt.title('Average Profit by Country')
plt.xlabel('Country')
plt.ylabel('Average Profit')
plt.xticks(rotation=45)
plt.show()

# Pie Chart for Sales by Product Category
sales_by_category = sales['Product_Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Product Category')
plt.show()
```
